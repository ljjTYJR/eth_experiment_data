```python
# Only compute success rate
# For draw tikz figures!
import os
import numpy as np
import math
poses = [
    " easyPoses",
    " mediumPoses",
    " hardPoses"
]
overlaps = [
    "high",
    "medium",
    "low"
]

# Use: add the name of txt files
# must be the `apartment`+'ratio"+'result'
files = [
    "gazebo5_result.txt",
    "gazebo25_result.txt",
    "gazebo50_result.txt",
    "gazebo75_result.txt",
    "gazebo100_result.txt"
]
validation_path = "/media/shuo/T7/ETH/gazebo_winter/gazebo_validation.csv"

e_t = 0.1
e_R = 2.5 / 180 * math.pi

# poses:
def compute_pose_ratio():

    pres_translation = "pose"+"translation"+".bat"
    pres_rotation = "pose"+"rotation"+".bat"
    if os.path.isfile(pres_translation):
        os.remove(pres_translation)
    if os.path.isfile(pres_rotation):
        os.remove(pres_rotation)

    tmp = np.loadtxt(validation_path, str, delimiter=',', skiprows=1)
    validation = tmp[::10]
    result = {}
    # read the results
    for i in range(len(files)):
        result[i] = np.loadtxt(files[i], float, delimiter=',', skiprows=0)

    # follow the order of ratio to compute different levels of results
    with open(pres_translation, 'a') as f:
        f.write("ratio" + " " + "easy" + " " + "medium" + " " + "hard")
        for i in range(len(files)):
            f.write("\n")
            # NOTE: gazebo!
            ratio = files[i][len('gazebo') : files[i].index('_')]
            f.write(str(ratio) + " ")
            # different levels
            for pose in poses:
                # print(result[i].shape)
                result_tmp = result[i][np.array([validation[:, 1] == pose]).reshape(-1)]
                res_pose = sum(1 for val in result_tmp[:,1] if val <= e_t) / len(result_tmp)
                f.write(str(res_pose) + " ")
    with open(pres_rotation, 'a') as f:
        f.write("ratio" + " " + "easy" + " " + "medium" + " " + "hard")
        for i in range(len(files)):
            f.write("\n")
            ratio = files[i][len('gazebo') : files[i].index('_')]
            f.write(str(ratio) + " ")
            # different levels
            for pose in poses:
                # print(result[i].shape)
                result_tmp = result[i][np.array([validation[:, 1] == pose]).reshape(-1)]
                res_pose = sum(1 for val in result_tmp[:,2] if val <= e_R) / len(result_tmp)
                f.write(str(res_pose) + " ")

def compute_overlap_ratio():
    overlap_translation = "overlap"+"translation"+".bat"
    overlap_rotation = "overlap"+"rotation"+".bat"
    if os.path.isfile(overlap_translation):
        os.remove(overlap_translation)
    if os.path.isfile(overlap_rotation):
        os.remove(overlap_rotation)
    
    tmp = np.loadtxt(validation_path, str, delimiter=',', skiprows=1)
    validation = tmp[::10]
    result = {}
    # read the results
    for i in range(len(files)):
        result[i] = np.loadtxt(files[i], float, delimiter=',', skiprows=0)
    
    with open(overlap_translation, 'a') as f:
        f.write("ratio" + " " + "high" + " " + "medium" + " " + "low")
        for i in range(len(files)):
            f.write("\n")
            ratio = files[i][len('gazebo') : files[i].index('_')]
            f.write(str(ratio) + " ")
            # different levels
            for overlap in overlaps:
                result_tmp = result[i]
                validation_tmp = validation
                if overlap == overlaps[0]:
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) >= 0.75).reshape(-1)]
                elif overlap == overlaps[1]:
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) < 0.75).reshape(-1)]
                    validation_tmp = validation_tmp[np.array(validation_tmp[:, 0].astype(float) < 0.75).reshape(-1)]
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) >= 0.5).reshape(-1)]
                    validation_tmp = validation_tmp[np.array(validation_tmp[:, 0].astype(float) > 0.50).reshape(-1)]
                else:
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) <= 0.50).reshape(-1)]
                res_overlap = sum(1 for val in result_tmp[:,1] if val <= e_t) / len(result_tmp)
                f.write(str(res_overlap) + " ")

    with open(overlap_rotation, 'a') as f:
        f.write("ratio" + " " + "high" + " " + "medium" + " " + "low")
        for i in range(len(files)):
            result_tmp = result[i]
            f.write("\n")
            ratio = files[i][len('gazebo') : files[i].index('_')]
            f.write(str(ratio) + " ")
            # different levels
            for overlap in overlaps:
                result_tmp = result[i]
                validation_tmp = validation
                if overlap == overlaps[0]:
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) >= 0.75).reshape(-1)]
                elif overlap == overlaps[1]:
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) < 0.75).reshape(-1)]
                    validation_tmp = validation_tmp[np.array(validation_tmp[:, 0].astype(float) < 0.75).reshape(-1)]
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) >= 0.5).reshape(-1)]
                    validation_tmp = validation_tmp[np.array(validation_tmp[:, 0].astype(float) > 0.50).reshape(-1)]
                else:
                    result_tmp = result_tmp[np.array(validation_tmp[:, 0].astype(float) <= 0.50).reshape(-1)]
                res_overlap = sum(1 for val in result_tmp[:,2] if val <= e_R) / len(result_tmp)
                f.write(str(res_overlap) + " ")
compute_pose_ratio()
compute_overlap_ratio()
```
